### Transformation 1. Check for duplicates

The purpose for checking for duplicates is to ensure that we are not counting any observations more than once. This is to ensure accuracy of shark observation data.

```python
df.duplicated(subset=['Observation'])
```

### Transformation 2: Create separate DataFrame displaying shark species and observations per year.

Because I am looking directly at the population fluctuations of various species of sharks to observe whether their population trends align with trends of shark attack occurances, I do not need to consider variables such as season, region, habitat, management zone, latitude, longitude, temperature, or dissolved oxygen. Therefore I must create a separate dataframe displaying only the shark species I am observing, as well as the year in which the observation was made. 

```python
species1 = df[['Year', 'GreatHammerhead', 'Lemon', 'Sandbar', 'Tiger', 'GreatWhite', 'Bonnethead', 'CaribbeanReef', 'Dusky', 'ScallopedHammerhead', 'Spinner']]
species1.head(10)
```

### Transformation 3: Convert shark species common names to scientific names for the sake of forming a relationship with IUCN API source. 

Because my API source from IUCN lists species using their scientific names, I need to change the headings from common names to scientific names. This will allow me to match records and data later on using species names. 

```python
species1 = species1.rename(columns={"GreatHammerhead":"Sphyrna mokarran",
                       "Lemon": "Negaprion brevirostris",
                       "Sandbar": "Carcharhinus plumbeus",
                       "Tiger": "Galeocerdo cuvier",
                       "GreatWhite": "Carcharodon carcharias",
                       "Bonnethead": "Sphyrna tiburo",
                       "CaribbeanReef": "Carcharhinus perezii",
                       "Dusky": "Carcharhinus obscurus",
                       "ScallopedHammerhead": "Sphyrna lewini",
                       "Spinner": "Carcharhinus brevipinna"})

species1.head(20)
```

### Transformation 4. Combine 'Year's into singular row for each year, combine column values by SUM

To ease reading the data table, I can combine all observations for each species into one cell for each year. Therefore, I can combine all duplicate 'Year' entries by getting the sum of values for each corresponding shark observation. 

```python
agg_functions = {'Sphyrna mokarran':'sum', 'Negaprion brevirostris':'sum',
                'Carcharhinus plumbeus':'sum', 'Galeocerdo cuvier':'sum', 'Carcharodon carcharias':'sum',
                'Sphyrna tiburo':'sum', 'Carcharhinus perezii':'sum', 'Carcharhinus obscurus':'sum',
                'Sphyrna lewini':'sum', 'Carcharhinus brevipinna':'sum'}

species2 = species1.groupby(species1['Year']).aggregate(agg_functions)
species2
```

### Transformation 5. Delete 'Sphyrna lewini' due to lack of observations throughout the observation period between 2009 and 2021. 

After looking at the simplified data frame, I noticed there are no observations for 'Sphyrna lewini', or the scalloped hammerheads. Therefore, this isn't very helpful for my plans for analysis, so I can remove the column from the dataset to allow me to focus more on the species that were in fact observed. 

```python
sharks = species2.drop('Sphyrna lewini', axis=1)
sharks
```
