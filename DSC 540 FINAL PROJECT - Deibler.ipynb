### I have to adjust a few things on the datasets in order to make them able to be merged. 

```python
## Adjust ESA dataset

esa = esa.loc[esa['Common Name'].str.contains('shark')]
esa = esa.drop(columns = 'ESA Listing Date', axis = 1)
esa = esa.drop(columns = 'Entity Description', axis=1)
esa = esa.drop(columns = 'index', axis =1)
```

```python
esa.head()
```

```python
## Adjust attacks dataset
attacks = attacks.drop(columns='index', axis = 1)
attacks = attacks.rename(columns={'Species':'Scientific Name'})
```

```python
attacks.head(10)
```

```python
esa.set_index('Scientific Name', inplace = True)
attacks.set_index('Scientific Name', inplace = True)
df = attacks.join(esa, lsuffix = '_left', rsuffix='_right', how='outer').reset_index()
```

#### Visualization #2. Number of attacks per shark family

#### Separating df into smaller DataFrame for top 3 shark families involved in attacks - Carcharhinidae, Galeocerdonidae, and Lamnidae

```python
mylist = ['Carcharhinidae', 'Galeocerdonidae', 'Lamnidae']
data = df.loc[df['Family'].isin(mylist)]
data
```

#### Visualization #3.  ESA Status and Total Attacks of Top 3 Attacking Shark Families

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data1 = {'Scientific Name':(data['Scientific Name']),
       'ESA Listing Status':(data['ESA Listing Status']),
       'Total Attacks':(data['Total'])}

dataf = pd.DataFrame(data1)
dataf = dataf.sort_values('Total Attacks', ascending=False)

ax = sns.barplot(x='Scientific Name', y='Total Attacks', data = dataf,
                 hue='ESA Listing Status', dodge=False)
plt.title('Total Attacks by Sharks in Carcharhinidae, Lamnidae, and Galeocerdonidae')
plt.xticks(rotation=90)
plt.ylabel('Attack Frequency')
plt.legend(bbox_to_anchor=(1.05,1), loc='upper left', borderaxespad=0)
```

#### Visualization #4. Pie Chart displaying proportion of attacks in each ESA Listing Status

```python
agg_functions = {'Non-fatal Attacks':'sum', 'Fatal Attacks':'sum', 'Total':'sum'}

esagroups = data.groupby(data['ESA Listing Status']).aggregate(agg_functions)
esa_groups = esagroups.reset_index()
esa_groups
```

```python
df = esa_groups['Total']
keys = esa_groups['ESA Listing Status']

palette_color = sns.color_palette('bright')

plt.pie(df, labels=keys, colors=palette_color, autopct='%.0f%%')

plt.title('Proportion of Attacks by ESA Listing Status')
plt.show()
```
